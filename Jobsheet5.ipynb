```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/female.tiff')
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
img3 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
img_gamma = np.zeros(img3.shape, img3.dtype)

print(' Gamma Correction pada citra ') 
print('----------------------------------') 
try: 
 gamma = int(input('Masukkan nilai Gamma: ')) 
except ValueError: 
 print('Error, not a number')

for y in range(img3.shape[0]):
  for x in range(img3.shape[1]):
    for z in range(img3.shape[2]):
      img_gamma[y,x,z] = np.clip(255 * pow(img3[y,x,z]/255,(1/gamma)), 0, 255);

final_frame = cv.hconcat((img3, img_gamma))
cv2_imshow(final_frame)
```

```python
print(' Simulasi Image Depth ')
print('----------------------------------')

img_5 = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/female.tiff', cv.IMREAD_GRAYSCALE)
img_bit_depth = np.zeros(img_5.shape, img_5.dtype)

bit_depth=2
level = 255 / (pow(2,bit_depth)-1)

for y in range(img_5.shape[0]):
  img_bit_depth[y] = np.clip(np.round(img_5[y]/level)*level, 0, 255)

final_frame = cv.hconcat((img_5, img_bit_depth))
cv2_imshow(final_frame)
```

```python
import glob
from math import log10, sqrt
  
img_galaxy = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/ASSETS-PVCK/Images/noises/*.jpg'):
    n = cv.imread(img)
    cv_img.append(n)

# AVERAGE 5 CITRA

dst = cv_img[0]
noise_number = 4 
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(img_galaxy, after): 
    mse = np.mean((img_galaxy - after) ** 2) 
    if(mse == 0):
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse)) 
    return psnr

after = dst
psnr = PSNR(img_galaxy, after)

print('Nilai PSNR adalah',psnr,'(dB)')
galaxy_frame = cv.hconcat((img_galaxy, dst))
cv2_imshow(galaxy_frame)
```
