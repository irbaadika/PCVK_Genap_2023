```python
cascade = cv.CascadeClassifier('/content/drive/MyDrive/ASSETS-PVCK/Images/haarcascades/haarcascade_frontalface_alt.xml')

# JOKOWI

jokowi = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/jokowi.jpg')
jokowi_rgb = cv.cvtColor(jokowi, cv.COLOR_BGR2RGB)
jokowi2 = jokowi_rgb.copy()

roi_jokowi = cascade.detectMultiScale(jokowi2)
for(x,y,w,h) in roi_jokowi:
  cv.rectangle(jokowi2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(jokowi_rgb)
plt.subplot(122),plt.imshow(jokowi2)
plt.show()

# KARTINI

kartini = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/kartini.jpg',0)
kartini_rgb = cv.cvtColor(kartini, cv.COLOR_BGR2RGB)
kartini2 = kartini_rgb.copy()

roi_kartini = cascade.detectMultiScale(kartini2)
for(x,y,w,h) in roi_kartini:
  cv.rectangle(kartini2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(kartini_rgb,cmap='gray')
plt.subplot(122),plt.imshow(kartini2,cmap='gray')
plt.show()

# KUCING

kucing = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/mycat.jpg',0)
kucing_rgb = cv.cvtColor(kucing, cv.COLOR_BGR2RGB)
kucing2 = kucing_rgb.copy()

roi_kucing = cascade.detectMultiScale(kucing2)
for(x,y,w,h) in roi_kucing:
  cv.rectangle(kucing2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(kucing_rgb,cmap='gray')
plt.subplot(122),plt.imshow(kucing2,cmap='gray')
plt.show()

# MASKER

mask = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/mask.png',0)
mask_rgb = cv.cvtColor(mask, cv.COLOR_BGR2RGB)
mask2 = mask_rgb.copy()

roi_mask = cascade.detectMultiScale(mask2)
for(x,y,w,h) in roi_mask:
  cv.rectangle(mask2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(mask_rgb,cmap='gray')
plt.subplot(122),plt.imshow(mask2,cmap='gray')
plt.show()

# MICHAEL JORDAN

mjordan = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/mjordan.jpg',0)
mjordan_rgb = cv.cvtColor(mjordan, cv.COLOR_BGR2RGB)
mjordan2 = mjordan_rgb.copy()

roi_mjordan = cascade.detectMultiScale(mjordan2)
for(x,y,w,h) in roi_mjordan:
  cv.rectangle(mjordan2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(mjordan_rgb,cmap='gray')
plt.subplot(122),plt.imshow(mjordan2,cmap='gray')
plt.show()

# SOLVAY

solvayconf = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/solvayconf.jpg',0)
solvayconf_rgb = cv.cvtColor(solvayconf, cv.COLOR_BGR2RGB)
solvayconf2 = solvayconf_rgb.copy()

roi_solvayconf = cascade.detectMultiScale(solvayconf2)
for(x,y,w,h) in roi_solvayconf:
  cv.rectangle(solvayconf2, (x,y),(x+w,y+h), (0,255,0), 3)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(solvayconf_rgb,cmap='gray')
plt.subplot(122),plt.imshow(solvayconf2,cmap='gray')
plt.show()
```

```python
cascade = cv.CascadeClassifier('/content/drive/MyDrive/ASSETS-PVCK/Images/haarcascades/haarcascade_frontalface_alt.xml')

mj3 = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/mjordan.jpg')
smile = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/smile.png', cv.IMREAD_UNCHANGED)

mj3_gray = cv.cvtColor(mj3, cv.COLOR_BGR2GRAY)

faces = cascade.detectMultiScale(mj3_gray, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in faces:
    smile_resized = cv.resize(smile, (w, h))
    mj3[y:y+h, x:x+w] = cv.addWeighted(mj3[y:y+h, x:x+w], 1, smile_resized[:, :, 0:3], 1, 0)

plt.subplot(111), plt.imshow(cv.cvtColor(mj3, cv.COLOR_BGR2RGB))
plt.show()
```

```python
bounding_box = cv.imread('/content/drive/MyDrive/ASSETS-PVCK/Images/facedet/people.jpg')

bb_gray = cv.cvtColor(bounding_box, cv.COLOR_BGR2GRAY)

cascade = cv.CascadeClassifier('/content/drive/MyDrive/ASSETS-PVCK/Images/haarcascades/haarcascade_frontalface_alt.xml')
roi_bb = cascade.detectMultiScale(bb_gray, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in roi_bb:
    cv.rectangle(bounding_box, (x, y), (x+w, y+h), (0, 0, 0), 2)

    roi_gray = bb_gray[y:y+h, x:x+w]
    eye_cascade = cv.CascadeClassifier('/content/drive/MyDrive/ASSETS-PVCK/Images/haarcascades/haarcascade_eye.xml')
    eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=5)

    for (ex, ey, ew, eh) in eyes:
        cv.rectangle(bounding_box, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (0, 0, 255), 2)

        if ex + ew/2 > w/2:
            cv.rectangle(bounding_box, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (255, 255, 255), 2)

plt.figure(figsize=(15, 5))
plt.subplot(111), plt.imshow(cv.cvtColor(bounding_box, cv.COLOR_BGR2RGB))
plt.show()
```
